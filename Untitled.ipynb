```python
import os
import json
import datetime
```

```python
os.listdir("src")
```

```python
with open('src/spotify-10-07-2022.json') as json_file:
    spotify_data = json.load(json_file)

with open('src/location-10-07-2022.json') as json_file:
    location_data = json.load(json_file)

with open('src/browser-10-07-2022.json') as json_file:
    browser_data = json.load(json_file)
```

```python
# make new cards

count = 0
card_list = []
for elm in location_data['features']:
    if (elm['geometry']['type'] == "Point"):
        count += 1
        print(count, elm['properties']['name'])

        card_item = {}
        card_item['id'] = count
        card_item['title'] = elm['properties']['name']
        card_item['description'] = "TODO"
        card_item['time'] = elm['timespan_sec']
        card_item["simple_time"] = datetime.datetime.fromtimestamp(elm['timespan_sec']['begin'] / 1e3).strftime("%H:%M") + \
                    " - " + datetime.datetime.fromtimestamp(elm['timespan_sec']['end'] / 1e3).strftime("%H:%M")


        spotify_songs = [song for song in spotify_data["songs"] \
                            if song["time_sec"] < elm['timespan_sec']['end'] and  
                            song["time_sec"] > elm['timespan_sec']['begin'] ]
        # print([song['album'] for song in spotify_songs])
        card_item['spotify'] = spotify_songs


        web_hits = [hit for hit in browser_data["history"] \
                            if hit["starting_time"] < elm['timespan_sec']['end'] and  
                            hit["starting_time"] > elm['timespan_sec']['begin'] ]
        print("num web hits....", len(web_hits), "\n")
        card_item['browser'] = web_hits

        card_list.append(card_item)

```
